```python
students_marks = {
"Alice": {"Math": 85, "Physics": 90, "Chemistry": 78, "English": 88, "Biology": 92},
"Bob": {"Math": 75, "Physics": 80, "Chemistry": 85, "English": 70, "Biology": 76},
"Charlie": {"Math": 95, "Physics": 89, "Chemistry": 92, "English": 94, "Biology": 90},
"David": {"Math": 65, "Physics": 70, "Chemistry": 68, "English": 72, "Biology": 66},
"Eva": {"Math": 88, "Physics": 92, "Chemistry": 81, "English": 85, "Biology": 87}
}

# Printing the dictionary
for student, marks in students_marks.items():
  print(f"{student}: {marks}")
```

```python
# Calculate average marks for each student
for student, marks_dict in students_marks.items():
    total_marks = sum(marks_dict.values())
    num_subjects = len(marks_dict)
    average_marks = total_marks / num_subjects
    print(f"{student}'s Average Marks: {average_marks:.2f}")
```
