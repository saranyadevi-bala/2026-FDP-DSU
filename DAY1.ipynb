```python
student_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 65,
    "Eve": 95
}

print("Student Marks Dictionary:", student_marks)
```

```python
total_marks = sum(student_marks.values())
num_students = len(student_marks)

average_marks = total_marks / num_students

print(f"The average marks of the students is: {average_marks:.2f}")
```

```python
highest_scorer = max(student_marks, key=student_marks.get)
lowest_scorer = min(student_marks, key=student_marks.get)

print(f"Highest Scorer: {highest_scorer} with {student_marks[highest_scorer]} marks")
print(f"Lowest Scorer: {lowest_scorer} with {student_marks[lowest_scorer]} marks")
```

```python
print(f"Students scoring above the average of {average_marks:.2f}:")
for student, marks in student_marks.items():
    if marks > average_marks:
        print(f"- {student} ({marks} marks)")
```

```python
search_name = input("Enter the name of the student to search: ")

if search_name in student_marks:
    print(f"{search_name} scored {student_marks[search_name]} marks.")
else:
    print(f"Student '{search_name}' not found.")
```

```python
student_to_update = input("Enter the name of the student to update marks for: ")
if student_to_update in student_marks:
    try:
        new_mark = int(input(f"Enter new marks for {student_to_update}: "))
        if 0 <= new_mark <= 100: # Assuming marks are between 0 and 100
            student_marks[student_to_update] = new_mark
            print(f"Marks for {student_to_update} updated to {new_mark}.")

            # Recalculate average
            total_marks_revised = sum(student_marks.values())
            num_students_revised = len(student_marks)
            average_marks_revised = total_marks_revised / num_students_revised
            print(f"Revised Student Marks Dictionary: {student_marks}")
            print(f"Revised average marks: {average_marks_revised:.2f}")
        else:
            print("Invalid mark. Marks should be between 0 and 100.")
    except ValueError:
        print("Invalid input. Please enter a number for marks.")
else:
    print(f"Student '{student_to_update}' not found.")
```
